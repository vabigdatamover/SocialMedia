```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
city_df=pd.read_csv('city.csv')
city_df.head()
```

```python
city_df.set_index('geoName')[['Joe Biden','Donald Trump']].plot.bar()
plt.ylim([0,100])
plt.xlabel('Google City Trends')
plt.savefig("CityTrends.png", bbox_inches="tight")
plt.show()
```
